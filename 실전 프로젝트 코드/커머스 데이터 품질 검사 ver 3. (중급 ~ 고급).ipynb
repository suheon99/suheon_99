# 중급 2-1

- 구매한 아이템(order_items에서 파악)에는 유저와 상품 정보가 있습니다. 이 중 **event가 발생하지 않은 유저/상품** 쌍이 존재하는지 둘을 merge하여 집계하고 그 비율을 시각화합니다(pie chart)

```python
event
```

```python
event['product_id'] = event.apply(lambda row : row.uri.split('/')[-1] 
                                        if row.event_type == 'product' else np.NaN, axis=1)
event_df = event.copy()
event_df
```

```python
event_df.info()
```

```python
# user_id가 결측치면 행을 삭제
event_df = event_df.dropna(subset=['user_id'])



# 'user_id'를 정수로 변환
event_df['user_id'] = event_df['user_id'].astype(int)
# user_id와 product_id 쌍을 합하여 check라는 열에 저장
event_df['check'] = event_df['user_id'].astype(str) + '/' + event_df['product_id']

event_df
```

```python
event_df.info()
```

```python
order.info()
```

```python
# detail_orders에도 check 컬럼 생성
order['check'] = order['user_id'].astype(str) + '/' + order['product_id'].astype(str)
order_df = order.copy()
order_df
```

```python
# check열 기준으로 left join. suffixes : 중복되는 열 이름 있으면 접미사 붙여서 구분하기
result = pd.merge(order_df, event_df, on='check', how='left', suffixes=('_order', '_event'))
result
```

```python
# order와 event df에서 각각 created_at이 가장 작은 거 택하기
order_grouped = order_df.groupby(['user_id', 'product_id'], as_index=False)['created_at'].min()
event_grouped = event_df.groupby(['user_id', 'product_id'], as_index=False)['created_at'].min()
```

```python
order_grouped
```

```python
event_grouped
```

```python
# 문자열로 통일하기
order_df['user_id'] = order_df['user_id'].astype(str)
order_df['product_id'] = order_df['product_id'].astype(str)

event_df['user_id'] = event_df['user_id'].astype(str)
event_df['product_id'] = event_df['product_id'].astype(str)
```

```python
first_event = event_df.groupby(['user_id', 'product_id'])['created_at'].min().reset_index()
first_order = order_df.groupby(['user_id', 'product_id'])['created_at'].min().reset_index()

# 두 개 데이터프레임 merge하기
merged_df = pd.merge(first_event, first_order, on=['user_id', 'product_id'], suffixes=('_event', '_order'))
```

```python
# event가 order보다 더 빠른 경우 가져오기
time_check = (merged_df['created_at_event'] < merged_df['created_at_order']).sum()
time_check
```

```python
total_cnt = len(merged_df)
ratio= (time_check / total_cnt if total_cnt > 0 else 0) * 100

print(total_cnt)
print(ratio)
```

```python
print(event['created_at'].dtype)
```

```python
event['created_at'] = pd.to_datetime(event['created_at'], errors='coerce')
event.info()
```

```python
# 세션별로 정렬하고 시간 간격을 계산하여 최대 30분을 초과하는 경우만 골라냄
over_30m = event.sort_values(['session_id', 'created_at']).groupby('session_id').apply(lambda group: group['created_at'].diff().gt(pd.Timedelta(minutes=30)).any()).sum()
over_30m
```

```python
# 전체 event 행으로 나눠서 비율 계산하기
ratio = over_30m / len(event)
ratio
```

```python
ratio_uni = over_30m / len(event['session_id'].unique())
ratio_uni
```

```python
event_df = event.sort_values(['session_id', 'created_at'])
event_df['session_check'] = event_df['session_id'] == event_df['session_id'].shift(-1)
event_df['time_diff'] = (event_df['created_at'].shift(-1) - event_df['created_at']).dt.total_seconds() / 60 >= 30
event_df['3_1_check'] = event_df['session_check'] & event_df['time_diff']
```
