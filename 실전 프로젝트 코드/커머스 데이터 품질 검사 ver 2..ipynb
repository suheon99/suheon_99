- 구매한 아이템(order_items에서 파악)에는 유저와 상품 정보가 있습니다. 이 중 **event가 발생하지 않은 유저/상품** 쌍이 존재하는지 둘을 merge하여 집계하고 그 비율을 시각화합니다(pie chart)

```python
order
```

```python
event
```

```python
merged_df = pd.merge(order, event, on='user_id', how='left', suffixes=('_order', '_event')) 
# 충돌하는 열의 이름을 구분하기 위해 각 열에 '_order' 또는 '_event'를 추가 

merged_df
```

```python
merged_df.columns
```

```python
missing_events_df = merged_df[merged_df['event_type'].isna()]

# 'event'가 발생하지 않은 유저/상품 쌍을 출력
print(missing_events_df[['user_id', 'product_id']])
```

```python
# 'event_type'이 NaN도 아니고 빈 값도 아닌 행을 찾음
valid_events_df = merged_df[merged_df['event_type'].notna()]


print(valid_events_df)
```

```python
missing_events_df.to_csv('missing_events.csv',index=False)
```

```python
valid_events_df.to_csv('valid_events.csv', index=False)
```
