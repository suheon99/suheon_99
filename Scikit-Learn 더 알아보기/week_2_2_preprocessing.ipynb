# scikit-learn의 기본 메서드!

* fit: 데이터를 입력으로 받아서 모델을 학습 시킴. train set에만 사용.
* transform: fit을 통해 학습된 구조를 사용해 데이터를 변환. test set에 사용.
* fit_transform: tit과 transform을 연속적으로 수행하는 편리한 방법. train set에만 사용.

## 저분산인 feature 제거

````python
df1['test'] = 100
df1
# 모든 값이 100인 컬럼
````

````python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()
transformed_df = selector.fit_transform(df1)  # numpy.ndarray 형태.
df1 = pd.DataFrame(transformed_df, columns=selector.get_feature_names_out()) # 데이터프레임 형태로 바꾸기
df1
# 100개 중 90개가 똑같은 컬럼 있다면 이렇게 없애면 됨.
````

* 모든 값이 같았던 'test'컬럼이 삭제됨.
* parameter 로 threshold 지정 가능. 만약 threshold=0.2 일 경우, 80% 이상이 동일한 값이 컬럼 삭제됨.
* get_feature_names_out() Method를 이용해 남아있는 데이터의 컬럼을 알 수 있음.
* 필요없는 컬럼이 있는 거 같다면 이런 기능을 사용하기
