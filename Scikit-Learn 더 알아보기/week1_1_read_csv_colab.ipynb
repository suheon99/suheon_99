### parse_dates

```python
# 첫 번째 열을 날짜 형태의 인덱스로 사용
df = pd.read_csv("/content/week1_1.csv", index_col=0, parse_dates=True)  # parse_dates=True면 index를 dates 형태로 파싱.
df.index.name = None  # index name 을 없애고 싶을 때.
print (df.index.dtype)
df.head()
```
