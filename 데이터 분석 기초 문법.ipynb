# Groupby
- 집단 별로 묶어서 통계치를 본다.
- 반별로의 수학평균, 영어평균 등
- groupby + agg 조합을 많이 씀.
- 데이터를 묶으면서, 내가 원하는 통계치를 얻을 수 있으니까

```python
df_sp.groupby('nclass').agg(mean_math = ('math','mean'))
# 변수와 통계치를 써 줌
```

```python
df_sp.groupby('nclass').agg(sum_math = ('math','sum'))
```

```python
df_sp.groupby('nclass').agg(max_math = ('math','max'))
```

```python
df_sp.groupby('nclass').agg(mean_math = ('math','mean'),
                            sum_science = ('science','sum'),
                            nclass_count = ('nclass','count'))
```
