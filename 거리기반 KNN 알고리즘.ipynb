```python
import mglearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('Raisin_Dataset.xlsx')
df
```

```python
df['Class'].value_counts()
```

## 독립변수, 종속변수 나누기

```python
x = df.drop(['Area','Class'],axis='columns')
y = df[['Class']]
```

```python
x
```

```python
y
```

# 거리기반 KNN 알고리즘

```python
from sklearn.preprocessing import StandardScaler

StandardScaler = StandardScaler()
StandardScaler
```

### **데이터 스케일링**

```python
df_mm = StandardScaler.fit_transform(x)
df_mm
# 스케일링 값
```

- 스케일링 된 값으로 **데이터프레임** 다시 만들기

```python
x_mm = pd.DataFrame(data=df_mm,columns=x.columns)
x_mm
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_mm,y,train_size=0.7,random_state=111)
```

```python
X_train
```

# KNN 모델 생성하기
- **n_neighbors**, **weights**, **metric**

```python
# 모델링에 필요한 알고리즘 불러오기
cls = KNeighborsClassifier(n_neighbors=3, weights='uniform',metric='minkowski')

cls.fit(X_train,y_train)

print(cls.score(X_train, y_train), 'train의 정확도')
print(cls.score(X_test, y_test),'test의 정확도')
```

# 좋은 모델인가?

- train, test **둘 간의 간격이 최소화**되는 것이 좋은 모델로의 방향성
- test는 train보다 낮은 수치지만 상대적으로 높은 성능 보여주는 게 좋은 모델이다.

- 학습에 사용한 **데이터양**
- 하나의 모델(x), **다른 모델과도 비교**
- 실제 모델의 **튜닝에 대한 진행 유무** 확인
- **데이터 전처리 과정**도 확인
- **평가 지표 (ex.F1 score)들의 스코어** 확인
    - 이외에도.. 여러가지 고민해 봐야할 조건들이 많다.
