# 9. 각 카테고리에서 가장 비싼 제품과 가장 저렴한 제품의 정보(제품 이름, 가격)와 이들 간의 가격 차이를 조회하세요.

```python
import pandas as pd

# 예시 데이터 생성
data = {
    'ProductID': [1, 2, 3, 4, 5, 6],
    'ProductName': ['Apple iPhone 13', 'Samsung Galaxy S22', 'Logitech Mouse', 'Nike Running Shoes', 'Adidas Soccer Ball', 'Sony Headphones'],
    'Price': [999.99, 899.99, 29.99, 120.00, 25.00, 150.00],
    'Category': ['Electronics', 'Electronics', 'Accessories', 'Apparel', 'Apparel', 'Electronics']
}

products_df = pd.DataFrame(data)
products_df
```

```python
extremes_df = products_df.groupby('Category').agg(MaxPrice = ('Price','max'), MinPrice = ('Price','min')).reset_index()

max_price_info = pd.merge(products_df,extremes_df,left_on = ['Category','Price'], right_on = ['Category','MaxPrice'])
min_price_info = pd.merge(products_df,extremes_df,left_on = ['Category','Price'], right_on = ['Category','MinPrice'])

price_difference = max_price_info[['Category','ProductName','Price']].merge(min_price_info[['Category','ProductName','Price']], on = 'Category', suffixes = ('_Max','_Min'))
price_difference['PriceDifference'] = price_difference['Price_Max'] - price_difference['Price_Min']

price_difference[['Category','ProductName_Max','Price_Max','ProductName_Min','Price_Min','PriceDifference']]
```

# 14. 각 환자의 연간 방문 횟수를 기준으로 환자를 다음과 같이 등급을 분류하고, 환자 ID와 해당 등급을 조회합니다:

- **Gold**: 연간 방문 횟수가 5회 이상인 경우
- **Silver**: 연간 방문 횟수가 3회 이상 5회 미만인 경우
- **Bronze**: 연간 방문 횟수가 3회 미만인 경우


```python
import pandas as pd

# 예시 데이터 생성
patient_visits_data = {
    'VisitID': [1, 2, 3, 4],
    'PatientID': [200, 201, 200, 202],
    'VisitDate': ['2023-01-15', '2023-02-20', '2023-03-25', '2023-04-30'],
    'Department': ['Cardiology', 'Neurology', 'Cardiology', 'Orthopedics']
}

patient_visits_df = pd.DataFrame(patient_visits_data)
```

```python
patient_visits_df
```

```python
patient_visits_df['VisitDate'] = pd.to_datetime(patient_visits_df['VisitDate'])
patient_visits_df = patient_visits_df[patient_visits_df['VisitDate'].dt.year == 2023]

visit_counts = patient_visits_df.groupby('PatientID').size().reset_index(name='VisitCount')

def classify_grade(count):
    if count >= 5:
        return 'Gold'
    elif count >= 3:
        return 'Silver'
    else:
        return 'Bronze'
    
visit_counts['PatientGrade'] = visit_counts['VisitCount'].apply(classify_grade)

visit_counts[['PatientID','PatientGrade']]
```

# 15. 각 환자가 현재 연도 내에서 가장 많이 방문한 부서와 해당 부서의 마지막 방문 날짜를 조회합니다. 같은 방문 횟수인 경우, 가장 최근 방문 날짜가 늦은 부서를 선택합니다.

```python
import pandas as pd

# 예시 데이터 생성
patient_visits_data = {
    'VisitID': [1, 2, 3, 4],
    'PatientID': [200, 201, 200, 202],
    'VisitDate': ['2023-01-15', '2023-02-20', '2023-03-25', '2023-04-30'],
    'DeptID': [1, 2, 1, 3]
}

departments_data = {
    'DeptID': [1, 2, 3],
    'DeptName': ['Cardiology', 'Neurology', 'Orthopedics']
}

patient_visits_df = pd.DataFrame(patient_visits_data)
departments_df = pd.DataFrame(departments_data)
```

```python
patient_visits_df
```

```python
departments_df
```

```python
patient_visits_df['VisitDate'] = pd.to_datetime(patient_visits_df['VisitDate'])
patient_visits_df = patient_visits_df[patient_visits_df['VisitDate'].dt.year == 2023]


merged_df = pd.merge(patient_visits_df, departments_df, on='DeptID')


visit_counts = merged_df.groupby(['PatientID', 'DeptName']).agg(
    VisitCount=pd.NamedAgg(column='VisitID', aggfunc='count'),
    LastVisitDate=pd.NamedAgg(column='VisitDate', aggfunc='max')
).reset_index()


max_visits_per_patient = visit_counts.groupby('PatientID')['VisitCount'].transform('max')


visit_counts = visit_counts[visit_counts['VisitCount'] == max_visits_per_patient]
visit_counts.sort_values(by=['PatientID', 'LastVisitDate'], ascending=[True, False], inplace=True)


final_df = visit_counts.drop_duplicates(subset=['PatientID'], keep='first')

final_df[['PatientID', 'DeptName', 'LastVisitDate']]
```

# 19. 각 직원별로 가장 긴 연속 근무일 수를 조회합니다. 연속 근무일은 WorkDays 테이블의 WorkDate를 기준으로 판단하며, 어떤 직원이 연속된 날짜에 근무했는지를 기반으로 최대 연속 근무일 수를 계산해야 합니다.

```python
import pandas as pd
from datetime import timedelta

# 예시 데이터 생성
employees_data = {
    'EmployeeID': [1, 1, 1, 2, 2],
    'EmployeeName': ['Alice', 'Alice', 'Alice', 'Bob', 'Bob'],
    'WorkDayID': [1, 2, 3, 4, 5],
    'WorkDate': ['2023-01-01', '2023-01-02', '2023-01-04', '2023-01-01', '2023-01-02']
}

workdays_df = pd.DataFrame(employees_data)
workdays_df
```

```python
workdays_df['WorkDate'] = pd.to_datetime(workdays_df['WorkDate'])


workdays_df.sort_values(by=['EmployeeID', 'WorkDate'], inplace=True)
workdays_df['PrevWorkDate'] = workdays_df.groupby('EmployeeID')['WorkDate'].shift(1)
workdays_df['Diff'] = (workdays_df['WorkDate'] - workdays_df['PrevWorkDate']).dt.days


workdays_df['Group'] = ((workdays_df['Diff'] != 1) | (workdays_df['Diff'].isnull())).cumsum()


workdays_df['GroupSize'] = workdays_df.groupby(['EmployeeID', 'Group'])['WorkDate'].transform('count')


max_continuous_workdays = workdays_df.groupby('EmployeeID')['GroupSize'].max().reset_index(name='MaxContinuousWorkDays')
max_continuous_workdays
```
